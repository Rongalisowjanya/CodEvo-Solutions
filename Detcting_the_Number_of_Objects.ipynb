```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = '/content/bitss.jpg'
image = cv2.imread(image_path)
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
blurred = cv2.GaussianBlur(gray, (5, 5), 0)
```

```python
binary_image = cv2.adaptiveThreshold(blurred, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                     cv2.THRESH_BINARY_INV, 11, 2)
kernel = np.ones((3, 3), np.uint8)
binary_image = cv2.morphologyEx(binary_image, cv2.MORPH_CLOSE, kernel, iterations=2)
```

```python
contours, _ = cv2.findContours(binary_image, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
num_objects = len(contours)
print(f"Number of objects detected: {num_objects}")
```

```python
output_image = image.copy()
cv2.drawContours(output_image, contours, -1, (0, 255, 0), 2)
```

```python
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.show()
```

```python
plt.imshow(binary_image, cmap='gray')
plt.title('Binary Image')
plt.show()
```

```python
plt.imshow(cv2.cvtColor(output_image, cv2.COLOR_BGR2RGB))
plt.title('Image with Contours')
plt.show()
```
